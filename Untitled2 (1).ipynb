```python
# Inicializar lista vacía para almacenar estudiantes
estudiantes = []

# Definir el número de estudiantes a ingresar
num_estudiantes = 15  # Cambia este número según la cantidad de estudiantes que quieras ingresar

# Leer datos de estudiantes
for i in range(num_estudiantes):
    nombre = input(f"Introduce el nombre del estudiante {i + 1}: ")

    # Leer las tres calificaciones
    nota1 = float(input("Introduce la primera calificación: "))
    nota2 = float(input("Introduce la segunda calificación: "))
    nota3 = float(input("Introduce la tercera calificación: "))

    # Calcular nota final
    nota_final = (nota1 + nota2 + nota3) / 3

    # Determinar si el estudiante aprobó o reprobó
    if nota1 >= 70 and nota2 >= 70 and nota3 >= 70:
        estado = f"{nota_final:.2f}"  # Guardar el promedio si aprobó
    else:
        estado = "NA"  # Guardar NA si reprobó

    # Agregar estudiante a la lista
    estudiantes.append((nombre, nota1, nota2, nota3, estado))

# Imprimir encabezado del reporte
print("\nREPORTE DE CALIFICACIONES")
print("NOMBRE      NOTA1 NOTA2 NOTA3 NOTA FINAL")

# Imprimir datos de estudiantes
for estudiante in estudiantes:
    nombre, nota1, nota2, nota3, estado = estudiante
    print(f"{nombre:10} {nota1:5.2f} {nota2:5.2f} {nota3:5.2f} {estado:10}")

# Imprimir resumen
print(f"\nTOTAL {num_estudiantes} ESTUDIANTES")
print("Para aprobar el curso, debe tener 70 o más en cada una de las 3 calificaciones; la calificación final será el promedio. En caso de haber reprobado uno o más exámenes, la calificación será NA (no acreditado).")
```
